## 抓取網址

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
 
 
options = Options()
options.add_argument("--disable-notifications")  # 取消所有的alert彈出視窗
 
browser = webdriver.Chrome(
    ChromeDriverManager().install(), chrome_options=options)
 
browser.get("https://www.bnext.com.tw/articles")
 
 
soup = BeautifulSoup(browser.page_source, "html.parser")
 

elements = soup.find_all("div class", {"div_tab": "a href"}) #屬性太多會重複...還在找解套...
 
for element in elements:
   
    url = element.find_all(("a").find("href"))
    print(url)
page_next = browser.find_element_by_class_name("more_btn")
page_next.click()  # 點擊下一頁按鈕
```
